```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
ll = linear_model.LinearRegression(normalize=True)
```

```python
dataset = pd.read_csv('DATA (1).csv')
```

```python
X_values = dataset.iloc[:,0:4]
y_values = dataset.iloc[:,4]
```

```python
ll.fit(X_values,y_values)
```

```python
ll.coef_
```

```python
ll.predict([[1,2,3,4]])
```

```python
print(ll.intercept_)
```

```python
predictions = ll.predict(X_values)
```

```python
plt.scatter(y_values,predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_values, predictions))
print('MSE:', metrics.mean_squared_error(y_values, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_values, predictions)))
```
